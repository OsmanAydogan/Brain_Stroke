```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import KNNImputer
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Veri setini yükle
df = pd.read_csv("brain_stroke.csv")

# Eksik verileri 'None' ile işaretleyin (eğer gerekliyse)
df['smoking_status'] = df['smoking_status'].replace('Unknown', np.nan)

# KNN imputer nesnesini oluşturun
imputer = KNNImputer(n_neighbors=4)

# Kategorik verileri sayısal değerlere dönüştürmek için get_dummies kullanın
df_encoded = pd.get_dummies(df, columns=['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'], drop_first=True)

# Eğitim verilerini imputer ile doldurun
df_filled = imputer.fit_transform(df_encoded)

# Doldurulmuş verileri DataFrame'e geri dönüştürün
df_filled = pd.DataFrame(df_filled, columns=df_encoded.columns)

# KNN sınıflandırıcı
knn = KNeighborsClassifier(n_neighbors=23)

# Bağımsız değişkenler ve hedef değişken
X = df_filled.drop('stroke', axis=1)
y = df_filled['stroke']

# Modelin doğruluk yüzdesini hesaplayın
accuracy_scores = cross_val_score(knn, X, y, cv=5, scoring='accuracy')

# Ortalama doğruluk yüzdesini yazdırın
print(f"Modelin ortalama doğruluk yüzdesi: {accuracy_scores.mean() * 100:.2f}%")

df.head(30)

```

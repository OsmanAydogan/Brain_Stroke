### Nedenleri nelerdir? 
**Değiştiremeyeceğiniz risk faktörleri**

- **Yaş** - kişi yaşlandıkça inme geçirme olasılığı artar.
- **Cinsiyet** - erkeklerin inme geçirme olasılığı kadınlara göre daha yüksektir.
- **Irk** - Afrikalı Amerikalılar Kafkasyalılara göre iki kat daha fazla inme riskiyle karşı karşıyayken, Hispaniklerin daha genç yaşta inme geçirme olasılığı Hispanik olmayan Kafkasyalılara göre daha yüksektir. 

**Değiştirebileceğiniz risk faktörleri**

- **Yüksek kan basıncı (hipertansiyon)** - bu en baskın inme risk faktörüdür ve değiştirilmesi en kolay olanıdır. Kan basıncınızı düzenli olarak kontrol edin ve kontrol altında tutun.
- **Sigara içmek** - tütün kullanımı inme riskinizi iki katına çıkarır. Eğer sigara içiyorsanız, bırakın.
- **Kilo** - Aşırı kilolu olmak sizi yüksek kolesterol, yüksek tansiyon ve diyabete yatkın hale getirir ve bunların hepsi inme riskini artırır. Aşırı kiloluysanız, diyetinizi değiştirin ve yağlı yiyecek alımınızı sınırlayın.
- **Diyabet** - insanları inme ile sonuçlanabilecek kardiyovasküler hastalıklara karşı duyarlı hale getirir. Diyabetiniz varsa, iyi kontrol altında tutun.
- Geçirilmiş inme veya GİA - başka bir inme geçirme riskinizi artırır. Bazı ilaçlar düzenli olarak alındığında inme riskini azaltabilir.
- **Kalp hastalığı** - kalp rahatsızlıkları, özellikle atriyal fibrilasyon (düzensiz kalp atışı), inme riskini artırır. Bazı ilaçlar düzenli olarak alınırsa riski azaltabilir.

## AÇIKLAMA
>.info() metodu, boş olmayan değerlerin sayısı ve her sütunun veri türü dahil olmak üzere DataFrame'in kısa bir özetini almak için kullanılır.
Çıktı size sütun adları, her sütundaki boş olmayan değerlerin sayısı ve her sütunun veri türü gibi bilgiler sağlayacaktır. Bu özet, veri kümesinin yapısını ve bütünlüğünü anlamak için faydalıdır.

```python
def missing(df):
    missing_values = df.isnull().sum()
    missing_percent = (missing_values / len(df)) * 100
    return pd.concat([missing_values, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
missing(df)
```

## AÇIKLAMA
>Bu kodda, DataFrame'de eksik değer olup olmadığını kontrol etmek için **isnull()** yöntemi kullanıldık.Ardından **sum()** yöntemi ile her sütundaki eksik değerlerin sayısını bulmak için kullandık. **missing_values** değişkeni her sütun için eksik değerlerin sayısını saklarken, **missing_percent** değişkeni de sayıyı toplam satır sayısına bölüp 100 ile çarparak eksik değerlerin yüzdesini hesaplar. Son olarak, **pd.concat()** fonksiyonu **missing_values** ve **missing_percent** serilerini sütunlar boyunca (axis=1) birleştirmek ve keys parametresini kullanarak 'Missing_Number' ve 'Missing_Percent' sütun adlarını atadık. Missing fonksiyonunu DataFrame df ile çağırdığınızda, her sütun için eksik değerlerin sayısını ve yüzdesini içeren yeni bir DataFrame elde ettik.



```python
data=df.copy()
```

## AÇIKLAMA
>df üzerinde copy() yöntemi çağrıldık. Bu, df'nin derin bir kopyası olan yeni bir DataFrame nesnesi oluşturur, yani df ile aynı veri ve sütun yapısına sahiptir, ancak bellekte ayrı bir nesnedir. Kopya daha sonra değişken verilere atanır ve kopyalanan DataFrame ile orijinal df'den bağımsız olarak çalışmanıza olanak tanır. Bu, orijinal DataFrame'i değiştirmeden veriler üzerinde işlem yapmak istediğimiz için bu yola başvurduk.




```python
data.head()
```

```python
from sklearn.preprocessing import PowerTransformer
columns_to_transform = [ 'age', 'avg_glucose_level', 'bmi']
pt = PowerTransformer(method='yeo-johnson')
for column in columns_to_transform:
    pt.fit(data[[column]])
    transformed_data = pt.transform(data[[column]])
    data[column] = transformed_data
```

```python
print(data[['age', 'avg_glucose_level', 'bmi']].head())
```

## AÇIKLAMA
>sklearn.preprocessing modülünden PowerTransformer sınıfının **yeo_johnson metodunu** kullanarak veri dönüşümü yaptık. Amaç,  ‘age’, ‘avg_glucose_level’ ve ‘bmi’ sütunlarındaki verilerİ normal bir dağılıma sahip olacak şekilde dönüştürmekti. Sütunlardaki verilerin dağılımını daha normal (Gauss dağılımına daha yakın) bir hale getirdik.

```python
from sklearn.preprocessing import LabelEncoder
columns_to_encode = ['gender',  'ever_married','work_type', 'Residence_type','smoking_status']
for column in columns_to_encode:
    
    le = LabelEncoder()

   
    data[column] = le.fit_transform(data[column])
```
